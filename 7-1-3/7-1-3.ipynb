```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df_x1 = df[['longitude']]
```

```python
df_x1.head()
```

```python
df_y = df[["median_house_value"]]
```

```python
plt.scatter(df_x1, df_y)
```

```python
df_x = df.drop(labels=["median_house_value"], axis=1)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(df_x, df_y)
```

```python
print(model.coef_, model.intercept_)
```

```python
x_test = np.array([[4, 1, 6, 2, 8, 0, 1, -3]])
```

```python
y_test = model.predict(x_test)
```

```python
yhat = model.predict(df_x)
```

```python
plt.scatter(df_y, yhat)
```

```python
plt.scatter(df_x[["longitude"]], df_y)
plt.scatter(df_x[["longitude"]], yhat)
```

```python
model.score(df_x, df_y)
```
